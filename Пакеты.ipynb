```python
import numpy as np
import pandas as pd 
normal_sample = np.random.normal(loc = 0.0, scale = 1.0, size = 1000)
random_sample = np.random.random(size = 1000)
gamma_sample = np.random.gamma(2, size = 1000)

df = pd.DataFrame = ({
    'normal': normal_sample,
    'random': random_sample,
    'gamma': gamma_sample 
}) 
df
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

plt.figure() 
_ = plt.boxplot(df['normal'], whis = [0,100])
_ = plt.boxplot([df['random'],df['normal'], df['gamma']], whis = [0,100])
```
